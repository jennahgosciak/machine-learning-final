```python
#Append datafiles
fp = r"G:/.shortcut-targets-by-id/1hgVHDpl_iZ7es6sKW0QRlC9X6SF58wk7/Machine Learning 4 Cities Final Project/Data/"  # use your path
all_files = glob.glob(os.path.join(fp+"2_intermediate_data/census_acs/", "*.csv"))     # advisable to use os.path.join as this makes concatenation OS independent

df_from_each_file = (pd.read_csv(f) for f in all_files)
concatenated_acs   = pd.concat(df_from_each_file, ignore_index=True)
# doesn't create a list, nor does it append to one
```

```python
list(concatenated_acs.columns)
```

```python
#create proportion variables
concatenated_acs['pctmale']= (concatenated_acs['males'] / concatenated_acs['totpop'])*100
concatenated_acs['pct18under']= (concatenated_acs['pop18under'] / concatenated_acs['totpop'])*100
concatenated_acs['pct18_34']= (concatenated_acs['pop18_34'] / concatenated_acs['totpop'])*100
concatenated_acs['pct35_64']= (concatenated_acs['pop35_64'] / concatenated_acs['totpop'])*100
concatenated_acs['pct65plus']= (concatenated_acs['pop65plus'] / concatenated_acs['totpop'])*100
concatenated_acs['pcthispanic']= (concatenated_acs['pophispanic'] / concatenated_acs['totpop'])*100
concatenated_acs['pctnhwhite']= (concatenated_acs['popnhwhite'] / concatenated_acs['totpop'])*100
concatenated_acs['pctnhblack']= (concatenated_acs['popnhblack'] / concatenated_acs['totpop'])*100
concatenated_acs['pctnhaian']= (concatenated_acs['popnhaian'] / concatenated_acs['totpop'])*100
concatenated_acs['pctnhasian']= (concatenated_acs['popnhasian'] / concatenated_acs['totpop'])*100
concatenated_acs['pctnhhawainpi']= (concatenated_acs['popnhhawainpi'] / concatenated_acs['totpop'])*100
concatenated_acs['pctnhother']= (concatenated_acs['popnhother'] / concatenated_acs['totpop'])*100
concatenated_acs['pctnhtwoplus']= (concatenated_acs['popnhtwoplus'] / concatenated_acs['totpop'])*100

#proprotion of housing units owner and renter occupied
concatenated_acs['pctownerocch']= (concatenated_acs['housingoccupied'] / concatenated_acs['housing'])*100
concatenated_acs['pctrentocch']= (concatenated_acs['renterhousing'] / concatenated_acs['housing'])*100

#crowding as a proportion
concatenated_acs['crowding']= ((concatenated_acs['oc_1_onehalf']+concatenated_acs['oc_onehalf_2']+concatenated_acs['oc_2plus']) / concatenated_acs['housing'])*100
concatenated_acs['crowdingrenter']= ((concatenated_acs['roc_1_onehalf']+concatenated_acs['roc_onehalf_2']+concatenated_acs['roc_2plus']) / concatenated_acs['renterhousing'])*100

concatenated_acs.head()
```
