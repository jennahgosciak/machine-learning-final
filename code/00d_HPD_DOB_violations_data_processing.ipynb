```python
#violations per bbl
hpd_per_bbl= hpd_2012_2022_v2["BBL"].value_counts().reset_index().rename({"index":"BBL", "BBL":"num_hpd_vio"}, axis = 1)
hpd_per_bbl
```

```python
# b & c violations per bbl
violations = ['C','B']

violations_b_c= hpd_2012_2022_v2[hpd_2012_2022_v2.Class.isin(violations)]

print(violations_b_c)
```

```python
violations_b_c_per_bbl= violations_b_c["BBL"].value_counts().reset_index().rename({"index":"BBL", "BBL":"hpd_b_c_vio"}, axis = 1)

violations_b_c_per_bbl
```

```python
#fire code violations

#not sure if we want this as a separate variable
codes = [176,181,188, 213,222,483, 512, 513, 514, 515, 516, 517, 518, 519, 520, 521,522,523,524,525,695, 696, 697, 698, 699,812, 813, 814, 815, 816, 817, 818, 819, 820, 821, 822]

fire_codes = hpd_2012_2022_v2[hpd_2012_2022_v2.OrderNumber.isin(codes)]

fire_codes
```

```python
fire_vio_per_bbl= fire_codes["BBL"].value_counts().reset_index().rename({"index":"BBL", "BBL":"hpd_fire_vio"}, axis = 1)


fire_vio_per_bbl
```

```python
#MERGE OF ALL AGGREGATE HPD VARIABLES PER BBL

hpd_agg_final = pd.merge(pd.merge(hpd_per_bbl,fire_vio_per_bbl,how = 'left',on='BBL'),violations_b_c_per_bbl,how = 'left', on='BBL')

hpd_agg_final
                        
                        
                        
                        
```

```python
hpd_agg_final = hpd_agg_final.fillna(0)

hpd_agg_final
```

```python
#export

hpd_agg_final.to_csv('hpd_agg_final.csv')
```
